## 1.1) Importing text files

```python
anger_file = open("ANGER sentences.txt","r").readlines()
love_file = open("love.txt","r").readlines()
friendship_file = open("friendship.txt","r").readlines()
happiness_file = open("happiness.txt","r").readlines()
success_file = open("success.txt","r").readlines()
failure_file = open("failure.txt","r").readlines()
self_worth_file = open("self worth.txt","r").readlines()



# print('anger_file : \n',anger_file,end="\n")
# print('\nlove_file : \n',love_file,end="\n")
# print('\nfriendship_file : \n',friendship_file,end="\n")
# print('\nhappiness_file : \n',happiness_file,end="\n")
# print('\nsuccess_file : \n',success_file,end="\n")
# print('\nfailure _file : \n',failure_file,end="\n")
# print('\nself_worth_file : \n',self_worth_file,end="\n")

```

## 1.2) PROCESSING TEXT



### 1.2.1) REMOVING "\n"

```python
def line_remover(list_file):
    fl = []
    for sent in list_file:
        strg = ' '
        for word in sent.split():
            strg = strg + " " + word.rstrip("\n")
            
        fl.append(strg.lstrip())
    return fl

anger_file = line_remover(anger_file)
love_file = line_remover(love_file)
friendship_file = line_remover(friendship_file)
happiness_file = line_remover(happiness_file)
success_file = line_remover(success_file)
failure_file = line_remover(failure_file)
self_worth_file = line_remover(self_worth_file)

# print("ANGER : \n",anger_file,end="\n")
# print("\nLOVE : \n",love_file,end="\n")
# print("\nFRIENDSHIP : \n",friendship_file,end="\n")
# print("\nHAPPINESS : \n",happiness_file,end="\n")
# print("\nSUCCESS : \n",success_file,end="\n")
# print("\nFAILURE : \n",failure_file,end="\n")
# print("\nSELF WORTH : \n",self_worth_file,end="\n")
```

### 1.2.2) REMOVING NON ALPHANUMERIC CHARACTERS

```python
def remove_non_alpha_numeric_characters(sentence):
    new_sentence = ''
    for alphabet in sentence:
        if alphabet.isalpha() or alphabet == ' ':
            new_sentence += alphabet
    return new_sentence
```

```python
def preprocess_data(X):
    X = [data_point.lower() for data_point in X]
    X = [remove_non_alpha_numeric_characters(
        sentence) for sentence in X]
    X = [data_point.strip() for data_point in X]
    X = [re.sub(' +', ' ',
                data_point) for data_point in X]
    return X

anger_file = preprocess_data(anger_file)
love_file = preprocess_data(love_file)
friendship_file = preprocess_data(friendship_file)
happiness_file = preprocess_data(happiness_file)
success_file = preprocess_data(success_file)
failure_file = preprocess_data(failure_file)
self_worth_file = preprocess_data(self_worth_file)

# print("ANGER : \n",anger_file,end="\n")
# print("\nLOVE : \n",love_file,end="\n")
# print("\nFRIENDSHIP : \n",friendship_file,end="\n")
# print("\nHAPPINESS : \n",happiness_file,end="\n")
# print("\nSUCCESS : \n",success_file,end="\n")
# print("\nFAILURE : \n",failure_file,end="\n")
# print("\nSELF WORTH : \n",self_worth_file,end="\n")
```

### 1.2.3) CREATING ALL SENTENCES LIST

```python
all_sent_list = anger_file + love_file + friendship_file + happiness_file + success_file + failure_file + self_worth_file
print(len(all_sent_list))
all_sent_list
```

### 1.2.4) CREATING ALL SENTIMENT LIST

```python
sentiment_list = ["anger"]*len(anger_file) + ["love"]*len(love_file) + ["friendship"]*len(friendship_file) + ["happiness"]*len(happiness_file) + ["success"]*len(success_file) + ["failure"]*len(failure_file) + ["self_worth"]*len(self_worth_file)
print(len(sentiment_list))
sentiment_list
```

### 1.2.5) CREATING VOCABULARY LIST

```python
def vocab_maker(lst):
    vocabulary = set()
    for data_point in lst:
        for word in data_point.split(' '):
            vocabulary.add(word)
    
    return list(vocabulary)

vocabulary = vocab_maker(all_sent_list)
print(len(vocabulary))
vocabulary
```

### 1.2.6) CREATING DOCUMENT VECTOR

```python
all_sent_encoded = []

def encode_sentence(sentence):
    sentence = preprocess_data([sentence])[0]    # preprocess_data() takes list as argument and return list as output. hence we used [sentence] and [0] at end as function will return a list of len 1 and at 0 index we have our preprocessed text.
    # print(sentence)
    sentence_encoded = [0] * len(vocabulary)
    for i in range(len(vocabulary)):
        if vocabulary[i] in sentence.split(' '):
            sentence_encoded[i] = 1
    return sentence_encoded

all_sent_encoded = [encode_sentence(sentence) for sentence in all_sent_list]

print(len(all_sent_encoded))
print('all_sent_encoded :\n')
# print(all_sent_encoded)
```

```python
classes = list(set(sentiment_list))
# print(sentiment_list)
sentiment_encoded = []
for data_point in sentiment_list:
    data_point_encoded = [0] * len(classes)
    for i in range(len(classes)):
        if classes[i] == data_point:
            data_point_encoded[i] = 1
    sentiment_encoded.append(data_point_encoded)

print(len(sentiment_encoded))
print("\nsentiment_encoded :\n")
sentiment_encoded
```

## 1.3) CREATING TRAINING DATA AND TEST DATA

```python
all_sent_train = all_sent_encoded                             #X_train
sentiment_train = sentiment_encoded                       #Y_train
all_sent_test = all_sent_encoded                              #X_test
sentiment_test = sentiment_encoded                       #Y_test
```

## 1.4) MODEL TRAINING

```python
model = Sequential()
model.add(Dense(units=84, activation='sigmoid',input_dim=len(all_sent_train[0])))
model.add(Dense(units=len(sentiment_train[0]), activation='softmax'))

model.compile(loss=categorical_crossentropy,optimizer=SGD(lr=0.01,momentum=1.0, nesterov=True))

model.fit(np.array(all_sent_train), np.array(sentiment_train), epochs=4000, batch_size=64)
```

### 1.4.1) LIST DOWN PREDICTIONS

```python
predictions = [argmax(pred) for pred in model.predict(np.array(all_sent_test))]
```

```python
print(predictions)
print(len(predictions))
```

## 1.5) MODEL EVALUATION

```python
correct = 0
for i in range(len(predictions)):
    if predictions[i] == argmax(sentiment_test[i]):
        correct += 1

print ("Correct:", correct)
print ("Total:", len(predictions))
print("Accuracy : ",correct*100/len(predictions))
```
