# Machine Learning on Wine Quality Analysis

### i) Getting Dataset from Kaggle
### ii) Importing Crucial Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### iii) Importing Dataset

```python
df = pd.read_csv("wine.csv")
df.head()
```

```python
df.info()
```

### iv) Handling Missing Values 

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum() #There is no missing values in the Dataset
```

### v) Label Encoding

```python
df['quality'] = pd.factorize(df.quality)[0] # 1 for Good and 0 for Bad
```

```python
df.head(20)
```

### Performing Feature Selection Technique 

### Using Chi2 Test

```python
x= df.iloc[:,:-1]
y= df['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.15)
```

```python
from sklearn.feature_selection import chi2
f_p_values = chi2(x_train,y_train)   #p_values means minimum value is more relevant
f_p_values                           #f_score means maximum value is more relevant
```

```python
f_values=pd.Series(f_p_values[0]) 
f_values.index=x_train.columns         
f_values
```

```python
f_values.sort_values(ascending=False)
# total sulphur dioxide and free sulphur dioxide are the most relevant feature  
```

### Plotting Chi2 Curve

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
x = np.arange(0, 20, 0.001)                    #x-axis ranges from 0 to 20 with .001 steps
plt.plot(x, chi2.pdf(x, df=4), label='df: 4')  #define multiple Chi-square distributions
plt.plot(x, chi2.pdf(x, df=8), label='df: 8') 
plt.plot(x, chi2.pdf(x, df=12), label='df: 12') 
plt.legend(title = "Parameters")
plt.legend()
```

### Dropping Irrelevant Feature

```python
df.drop(['density','pH','sulphates','residual sugar','chlorides'],axis=1, inplace = True)
df.head()
```

### Pairplot 

```python
import seaborn as sns   # Defining the relationship between quality and other feature of the dataset 
sns.pairplot(df,hue ="quality")
```

### Naive Bayes
### For Model Evaluation

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train,y_train)
```

```python
classifier.score(x_train,y_train)
```

### Classification report

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,plot_confusion_matrix
from sklearn.datasets import make_classification
x, y = make_classification(random_state=0)
pred = classifier.predict(x_train)
```

```python
print(classification_report(y_train, pred))
print('Confusion Matrix: ', confusion_matrix(y_train, pred))
print('Accuracy: ', accuracy_score(y_train, pred))
```
